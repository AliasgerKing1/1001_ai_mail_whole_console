```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
dataset = pd.read_csv(r"D:\\datasets\\term_insuarance.csv")

dataset
```

```python
# Replace 'no' with 0 and 'yes' with 1 in the 'brought_insurance' column
dataset['brought_insurance'].replace({'no': 0, 'yes': 1}, inplace=True)
```

```python
# Assuming you have loaded your data into 'dataset' with columns 'age' and 'brought_insurance'
x = dataset['age']
y = dataset['brought_insurance']

# Calculate the best-fit line using numpy's polyfit function
coefficients = np.polyfit(x, y, 1)
slope = coefficients[0]
intercept = coefficients[1]

# Create a scatter plot
plt.scatter(x, y, label='Data Points')

# Add the best-fit line to the plot
plt.plot(x, slope * x + intercept, color='red', label='Best Fit Line')

# Add labels and a legend
plt.xlabel('Age')
plt.ylabel('Brought Insurance')
plt.legend()

# Show the plot
plt.show()
```
